```python
def paginador(nome,preço,total_paginas,cont_pag):
    n_paginas_feitas=0
    nav= abre_janela(cont_pag)
    try:
        for i in range (1,20): 
            time.sleep(1)
            if i==total_paginas:  
                for n in range (0,10):
                    try:
                        nome.append(str(nav.find_element_by_id(f'result_{n}_name').text))
                        preço.append(0.01*(int(nav.find_element_by_xpath(f'//*[@id="result_{n}"]/div[1]/div[2]/span[1]/span[1]').get_attribute('data-price'))))
                    except:
                        n_paginas_feitas+=1
                        break
                break
            else:
                for n in range (0,10):
                    nome.append(str(nav.find_element_by_id(f'result_{n}_name').text))
                    preço.append(0.01*(int(nav.find_element_by_xpath(f'//*[@id="result_{n}"]/div[1]/div[2]/span[1]/span[1]').get_attribute('data-price'))))
                n_paginas_feitas+=1
            nav.find_element_by_xpath('//*[@id="searchResults_btn_next"]').click()
        nav.quit()
        return n_paginas_feitas
    except:
        nav.quit()
        print('Erro de Busca na Página ')
        time.sleep(180)
        return n_paginas_feitas
        
```

```python
def abre_janela(contador_pag):
    nav = webdriver.Chrome()
    nav.get(f"https://steamcommunity.com/market/search?appid=730#p{contador_pag}_popular_desc")
    return nav
```
