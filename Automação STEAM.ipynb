```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd

navegador = webdriver.Chrome()
navegador.get("https://steamcommunity.com/market/search?appid=730")
total_paginas = int(navegador.find_element_by_xpath('//*[@id="searchResults_links"]/span[7]').text)

name =[] 
price = []
for i in range(1,5):
    time.sleep(1)
    for i in range (0,10):
        name.append(str(navegador.find_element_by_id(f'result_{i}_name').text))
        price.append(0.01*(int(navegador.find_element_by_xpath(f'//*[@id="result_{i}"]/div[1]/div[2]/span[1]/span[1]').get_attribute('data-price'))))
    navegador.find_element_by_xpath('//*[@id="searchResults_btn_next"]').click()
navegador.quit()

itens_df = pd.DataFrame(list(zip(name,price)),columns=['Nome','Preço USD$'])
# Get informations of CSGO itens from the steam website (name and price)
```
